## apply custom function on columns (Series Data)


```python
def grade(per: int) -> str:
    if per >= 80:
        return "A+"
    elif per >= 70:
        return "A"
    elif per >= 60:
        return "B"
    elif per >= 50:
        return "C"
    elif per >= 40:
        return "D"
    else:
        return "F"


grade(87)
```

### Multiple values

```python
def fn_obtain(s1: int, s2: int, s3: int, s4: int, s5: int) -> tuple[int, int, float, str]:
    total: int = 500
    obtained: int = s1 + s2 + s3 + s4 + s5
    per: float = obtained / total * 100
    grade: str = ""
    if per >= 80:
        grade = "A+"
    elif per >= 70:
        grade = "A"
    elif per >= 60:
        grade = "B"
    elif per >= 50:
        grade = "C"
    elif per >= 40:
        grade = "D"
    else:
        grade = "F"

    return total, obtained, per, grade


df1["total", "Obtained", "percentage", "grade"] = df1[["s1", "s2", "s3", "s4", "s5"]].apply(
    lambda x: fn_obtain(*x),
    axis=1,
    result_type="expand"
)
df1
```
