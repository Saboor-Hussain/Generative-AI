## apply custom function on columns (Series Data)

```python
def grade(per: int) -> str:
    if per >= 80:
        return "A+"
    elif per >= 70:
        return "A"
    elif per >= 60:
        return "B"
    elif per >= 50:
        return "C"
    elif per >= 40:
        return "D"
    else:
        return "F"
    
grade(87)
```

```python
df['percentage'].apply(grade)
```

```python
df['grade'] = df['percentage'].apply(grade)

df
```
